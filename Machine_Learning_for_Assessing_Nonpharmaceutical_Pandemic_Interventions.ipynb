```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Filter the world map to include only Africa
world_filtered = world[(world['continent'] == 'Africa')]

# Define countries to highlight and their respective colors
highlighted_countries = {'Nigeria': 'lightgreen',
                         'Ghana': 'coral',
                         'Mali': 'lightcyan',
                         'Burkina Faso': 'yellow',
                         'Gambia': 'red',
                         'Benin': 'seagreen',
                         "CÃ´te d'Ivoire": 'steelblue',
                         'Guinea': 'purple',
                         'Guinea-Bissau': 'lightgoldenrodyellow',
                         'Liberia': 'lightpink',
                         'Niger': 'lightblue',
                         'Sierra Leone': 'brown',
                         'Togo': 'magenta'}

# Plot the world map with increased size
fig, ax = plt.subplots(figsize=(12, 12))

# Plot outermost border with thicker line and lighter color
world_filtered.plot(ax=ax, color='white', edgecolor='black')

# Plot internal borders with thinner line and darker color
world_filtered.plot(ax=ax, color='white', edgecolor='lightgrey', linewidth=0.1, linestyle='--')

# Highlight selected countries with different colors
for country, color in highlighted_countries.items():
    country_data = world_filtered[world_filtered.name == country]
    if not country_data.empty:
        country_data.plot(ax=ax, color=color, edgecolor='black')

        # Add country name annotation
        centroid = country_data.geometry.centroid.values[0]
        ax.text(centroid.x, centroid.y, country, fontsize=7, ha='center', fontweight='bold')

# Remove axis spines and ticks
ax.set_xticks([])
ax.set_yticks([])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.tick_params(bottom=False, left=False)

# Set aspect ratio to 'equal'
ax.set_aspect('equal')

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_correlation_heatmap(ax, data, target_column, title=''):
    corr = data.corr()
    target_corr = corr[target_column].drop(target_column)
    target_corr_sorted = target_corr.sort_values(ascending=False)

    sns.set(font_scale=0.8)
    sns.set_style("white")
    sns.set_palette("PuBuGn_d")
    sns.heatmap(target_corr_sorted.to_frame(), cmap="plasma", annot=True, fmt='.2f', ax=ax)
    ax.set_title(title, fontsize=10, fontweight='bold')

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot correlation heatmap for new_cases_smoothed
plot_correlation_heatmap(axes[0], clean_varcase, 'new_cases_smoothed', title='(A) New cases')

# Plot correlation heatmap for new_deaths_smoothed
plot_correlation_heatmap(axes[1], clean_vardeath, 'new_deaths_smoothed', title='(B) New deaths')

# Plot correlation heatmap for positive_rate
plot_correlation_heatmap(axes[2], clean_varpositive, 'positive_rate', title='(C) Positive rate')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import silhouette_score

# Sample data (replace with your dataset)
X = cleaned_df[features]

# Range of clusters to try
min_clusters = 2
max_clusters = 10

silhouette_scores = []

# Iterate over different numbers of clusters
for n_clusters in range(min_clusters, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot silhouette scores
plt.plot(range(min_clusters, max_clusters + 1), silhouette_scores, marker='o')
plt.xlabel('Number of clusters (K)', fontweight='bold', fontsize=11)
plt.ylabel('Silhouette score', fontweight='bold', fontsize=11)
plt.title('Silhouette Method for Optimal K', fontweight='bold')
plt.show()

# Find the optimal number of clusters with the highest silhouette score
optimal_clusters = np.argmax(silhouette_scores) + min_clusters
print("Optimal number of clusters:", optimal_clusters)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have already clustered your data and have labels in numeric_df["cluster"]

# Define the main clusters
main_cluster_1 = 0  # Adjust this to the label of your first main cluster
main_cluster_2 = 1  # Adjust this to the label of your second main cluster
main_cluster_3 = 2
main_cluster_4 = 3

# Calculate the centroids of the three main clusters
centroid_1 = np.mean(reduced_data[numeric_df["cluster"] == main_cluster_1], axis=0)
centroid_2 = np.mean(reduced_data[numeric_df["cluster"] == main_cluster_2], axis=0)
centroid_3 = np.mean(reduced_data[numeric_df["cluster"] == main_cluster_3], axis=0)

# Check if any of the clusters are empty
if np.isnan(centroid_1).any() or np.isnan(centroid_2).any() or np.isnan(centroid_3).any():
    print("One or both of the main clusters are empty. Adjust cluster labels.")
else:
    # Calculate the distances from each point to the centroids
    distances_1 = np.linalg.norm(reduced_data[numeric_df["cluster"] == main_cluster_1] - centroid_1, axis=1)
    distances_2 = np.linalg.norm(reduced_data[numeric_df["cluster"] == main_cluster_2] - centroid_2, axis=1)
    distances_3 = np.linalg.norm(reduced_data[numeric_df["cluster"] == main_cluster_3] - centroid_3, axis=1)

    # Determine the radius of the circle (e.g., as the maximum distance from the centroid)
    radius_1 = np.max(distances_1)
    radius_2 = np.max(distances_2)
    radius_3 = np.max(distances_3)

    # Plot clustered data
    plt.figure(figsize=(8, 8))
    for cluster in range(9):
        plt.scatter(reduced_data[numeric_df["cluster"] == cluster, 0],
                    reduced_data[numeric_df["cluster"] == cluster, 1],
                    label=feature_labels[cluster],
                    color=colors[cluster])

    # Plot boundary circles
    circle_1 = plt.Circle(centroid_1[:2], radius_1, color='gray', fill=False)
    circle_2 = plt.Circle(centroid_2[:2], radius_2, color='gray', fill=False)
    circle_3 = plt.Circle(centroid_3[:2], radius_3, color='gray', fill=False)

    plt.gca().add_patch(circle_1)
    plt.gca().add_patch(circle_2)
    plt.gca().add_patch(circle_3)

    # Calculate the centroids of the three main clusters
centroid_4 = np.mean(reduced_data[numeric_df["cluster"] == main_cluster_4], axis=0)

# Check if the cluster is empty
if np.isnan(centroid_4).any():
    print("The fourth main cluster is empty. Adjust cluster labels.")
else:
    # Calculate the distances from each point to the centroid
    distances_4 = np.linalg.norm(reduced_data[numeric_df["cluster"] == main_cluster_4] - centroid_4, axis=1)

    # Determine the radius of the circle (e.g., as the maximum distance from the centroid)
    radius_4 = np.max(distances_4)

    # Plot boundary circle
    circle_4 = plt.Circle(centroid_4[:2], radius_4, color='gray', fill=False)
    plt.gca().add_patch(circle_4)


    plt.title('', fontsize=14, fontweight='bold')
    plt.xlabel('Principal Component 1', fontsize=12, fontweight='bold')
    plt.ylabel('Principal Component 2', fontsize=12, fontweight='bold')
    plt.legend()
    plt.grid(False)
    plt.show()

```
